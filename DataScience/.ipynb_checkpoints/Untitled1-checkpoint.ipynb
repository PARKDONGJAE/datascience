```python
def countMissing(v):
    nullCount = np.sum(pd.isnull(v))
    return nullCount
def propMissing(v):
    n=countMissing(v)
    return n/v.size
```

```python
# titanic.apply(countMissing)
titanic.apply(propMissing, axis =1)
```
