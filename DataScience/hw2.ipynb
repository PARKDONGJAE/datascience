## toji.txt 대화부분("")추출


```python
import re
from konlpy.tag import Hannanum, Kkma, Komoran, Okt
txt = open("../toji.txt", encoding='utf-16').read()
word = re.findall('"[^a-zA-Z].+"',txt)
word
list = []
for i in range(len(word)):
    list.append(word[i].strip("\""))
final = ",".join(list)
okt = Okt()
okt.nouns(final)
```
